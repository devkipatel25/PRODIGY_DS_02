```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_movies.txt')
```

```python
df
```

```python
table = (df[(df["Popularity"] >= 1000) & (df["Release_Date"] >= "2021-01-01")])
# table['Release_Date'] = pd.to_datetime(table['Release_Date'])
# table = table.sort_values(by='Release_Date', ascending=True)
```

```python
popularity = [float(value) for value in table["Popularity"]]
# popularity
```

```python
releasedate = [str(value) for value in table["Release_Date"]]
# releasedate
```

```python
plt.scatter(popularity, releasedate)
plt.title('Popularity vs Release Date')
plt.xlabel('Popularity')
plt.ylabel('Release Date')
plt.grid(True)
plt.show()
```

```python
genres = df['Genre'].str.split(',').explode().str.strip().str.lower()
# genres
```

```python
counts = genres.value_counts().head(6)
# counts
```

```python
genre_df = pd.DataFrame({'Genre': counts.index, 'Count': counts.values})
genre_df
```

```python
sns.barplot(x='Genre', y='Count', data=genre_df, palette='viridis')
plt.xticks(rotation=30)
plt.title('Distribution of Genres')
plt.tight_layout()
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```
